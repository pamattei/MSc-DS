# MSc Data Science: intro to HD regression: Ridge and Moore-Penrose

First we install the ppls package, where we will take the cookie data set. The features are NIR spectra from cookies, and we have 4 responses: the percentages of fat, sucrose, dry flour, and water.

```r
install.packages("ppls")
```

```r
library("ppls")
```

```r
data(cookie)
cookie = scale(cookie)

head(cookie)
```

Create a training and test sets, and fit a Moore-Penrose regression model, and a several ridge regression models with different hyperparameters, for example to predict the percentage of fat in the cookie. Discuss the results.

```r
X = cookie[,1:700]
Y = cookie[,701]
```

```r
Y # fat content of cookies
```

```r
plot(X[2,],type="l",main = "Spectra of a cookie")
```

```r
n = dim(cookie)[1]
p = dim(X)[2]
ind_test = sample(n,15) # The test set will contain 15 cookies

Xtrain = X[-ind_test,]
Xtest = X[ind_test,]
Ytrain = Y[-ind_test]
Ytest = Y[ind_test]
```

We can check that $X^TX$ is not invertible by computing its determinant:

```r
det(t(Xtrain)%*%Xtrain)
```

... which is zero because $p$ is larger than $n$.

The R function for Moore-Penrose pseudo-inversion is $\texttt{ginv}$. Let's compute $\beta_{MP}$ now using it:

```r
beta_MP = ginv(t(Xtrain)%*%Xtrain) %*%t(Xtrain) %*% Ytrain
```

Let's check that the training error is indeed zero (or very very close to zero):

```r
mean((Xtrain %*% beta_MP - Ytrain)^2)
```

Let's now predict the value of the fat content of the cookies in the test set, and look at the test MSE:

```r
Y_pred_MP = Xtest %*% beta_MP
print(mean((Y_pred_MP - Ytest)^2))
```

Let's now implement ridge regression:

```r
lambda = 0.1

beta_r = solve(t(Xtrain)%*%Xtrain + lambda*diag(p)) %*% t(Xtrain) %*% Ytrain
```

```r
mean((Xtrain %*% beta_r - Ytrain)^2)
```

We can see that the training error is bigger than for the Moore-Penrose estimate. Increasing $\lambda$ leads to bigger training error, but potientially less overfitting.

Let's now look at the test error:

```r
Y_pred_r = Xtest %*% beta_r
print(mean((Y_pred_r - Ytest)^2))
```

For $\lambda = 0.1$, the test error or ridge is better than the one of Moore-Penrose.

```r
lambda = 100

beta_r = solve(t(Xtrain)%*%Xtrain + lambda*diag(p)) %*% t(Xtrain) %*% Ytrain

Y_pred_r = Xtest %*% beta_r
print(mean((Y_pred_r - Ytest)^2))
```

For $\lambda = 100$, ridge's test error is much worse than the one of Moore-Penrose.
